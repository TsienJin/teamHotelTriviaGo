```python
import PyPDF2
```

```python
file = open("sample.pdf", "rb")
```

```python
reader = PyPDF2.PdfFileReader(file)
```

```python
print(reader.numPages)
```

```python
page7 = reader.getPage(6)
pdfData7 = page7.extractText()
print(pdfData7)
```

```python
for x in range(reader.numPages):
    page = reader.getPage(x)
    pdfData = page.extractText()
    print(pdfData)
```
